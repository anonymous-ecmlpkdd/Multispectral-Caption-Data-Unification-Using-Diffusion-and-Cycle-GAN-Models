```python
import os
from PIL import Image, ImageOps

src_dir = "/content/images2"
dst_dir = "/content/images2_resized"

if not os.path.exists(dst_dir):
    os.makedirs(dst_dir)

for fname in os.listdir(src_dir):
    if fname.lower().endswith('.jpg'):
        src_path = os.path.join(src_dir, fname)
        dst_path = os.path.join(dst_dir, fname)
        try:
            with Image.open(src_path) as img:
                img = img.convert("RGB")

                img = ImageOps.fit(img, (299, 299))
                img.save(dst_path, quality=95)
        except Exception as e:
            print(f"{src_path} dönüştürürken hata: {e}")

print("Ön işleme tamamlandı. Yeni dizindeki dosya sayısı:", len(os.listdir(dst_dir)))
```

```python
import os
from PIL import Image, ImageOps

src_dir = "/content/generated_images"
dst_dir = "/content/generated_images_resized"

if not os.path.exists(dst_dir):
    os.makedirs(dst_dir)

# Tüm .jpg dosyalarını dönüştürüp kaydediyoruz
for fname in os.listdir(src_dir):
    if fname.lower().endswith('.jpg'):
        src_path = os.path.join(src_dir, fname)
        dst_path = os.path.join(dst_dir, fname)
        try:
            with Image.open(src_path) as img:
                img = img.convert("RGB")
                # ImageOps.fit, görüntüyü merkezden kırparak tam olarak (299,299) üretir
                img = ImageOps.fit(img, (299, 299))
                img.save(dst_path, quality=95)
        except Exception as e:
            print(f"{src_path} dönüştürürken hata: {e}")

print("Ön işleme tamamlandı. Yeni dizindeki dosya sayısı:", len(os.listdir(dst_dir)))
```

```python
from torch_fidelity import calculate_metrics

real_images_dir = "/content/images2_resized"
generated_images_dir = "/content/generated_images_resized"

metrics = calculate_metrics(
    input1=real_images_dir,    
    input2=generated_images_dir,
    cuda=True,
    isc=True,
    fid=True,
    kid=True,
    num_workers=0,
    batch_size=32,
)

print("FID:", metrics.get("fid"))
print("Inception Score:", metrics.get("isc"))
print("KID:", metrics.get("kid"))
```

```python
print("Frechet Inception Distance:", metrics.get("frechet_inception_distance"))
print("Inception Score:", metrics.get("inception_score_mean"), "±", metrics.get("inception_score_std"))
print("Kernel Inception Distance:", metrics.get("kernel_inception_distance_mean"), "±", metrics.get("kernel_inception_distance_std"))
```
